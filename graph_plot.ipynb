```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    data = pd.read_csv("stat_data.csv")
    print(data.describe())
    fig,ax = plt.subplots(2,2)
    fig.suptitle("Http rqeuest GET and POST round trip graph")
    ax[0,0].set_title("For 2:1 request ratio")
    ax[0,0].plot(data["client_count"],data["2:1 req"])
    ax[0,1].set_title("For 5:1 request ratio")
    ax[0,1].plot(data["client_count"],data["5:1 req"])
    ax[1,0].set_title("For 10:1 request ratio")
    ax[1,0].plot(data["client_count"],data["10:1 req"])
    

    plt.tight_layout()
    fig.delaxes(ax[1][1])
    plt.show()
    fig.savefig("stat_data_fig.png")
    
```

```python
if __name__ == "__main__":
    main()
```
